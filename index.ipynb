```python
# Assuming 'sucide' is  DataFrame with columns: 'country', 'year', 'sex', and 'suicides_no'
# Filter the data
df1 = sucide[sucide.country == "Israel"]
df2 = df1[df1.year >= 2000]
df_male = df2[df2.sex == 'male']
df_female = df2[df2.sex == 'female']

# Group by year and sum the suicides
male_suicides = df_male.groupby('year')['suicides_no'].sum()
female_suicides = df_female.groupby('year')['suicides_no'].sum()

# Create the bar chart
fig, ax = plt.subplots(figsize=(12, 7))
index = male_suicides.index
bar_width = 0.35

bar1 = plt.bar(index, male_suicides, bar_width, label='Male Suicides')
bar2 = plt.bar(index + bar_width, female_suicides, bar_width, label='Female Suicides')

plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.title('Male vs Female Suicides in Israel (2000-2015)')
plt.grid()
plt.xticks(index + bar_width / 2, index)
plt.legend()

plt.tight_layout()
plt.show()
```
